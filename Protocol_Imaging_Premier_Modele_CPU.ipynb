````python
### CELLULE 7: Cr√©er le mapping (Images ‚Üî Protocoles)
# =============================================================================

print("\n" + "="*80)
print("CR√âATION DU MAPPING")
print("="*80)

import re

# Fonction pour extraire le num√©ro du nom de fichier
def extract_number_from_filename(filename):
    """Extraire le premier nombre trouv√© dans le nom de fichier"""
    match = re.search(r'(\d+)', filename)
    if match:
        return int(match.group(1))
    return 0

# Strat√©gie A: Nom de fichier ‚Üí num√©ro de requ√™te
ocr_results['requete_from_filename'] = ocr_results['image_name'].apply(extract_number_from_filename)

data_a = ocr_results.merge(
    protocoles,
    left_on='requete_from_filename',
    right_on='Requete ',
    how='inner'
)

# Strat√©gie B: Num√©ro de carte ‚Üí num√©ro de requ√™te
ocr_results['card_as_int'] = pd.to_numeric(ocr_results['card_number'], errors='coerce')

data_b = ocr_results.merge(
    protocoles,
    left_on='card_as_int',
    right_on='Requete ',
    how='inner'
)

# Strat√©gie C: Num√©ro de permis ‚Üí num√©ro de requ√™te
ocr_results['permis_as_int'] = pd.to_numeric(ocr_results['permis'], errors='coerce')

data_c = ocr_results.merge(
    protocoles,
    left_on='permis_as_int',
    right_on='Requete ',
    how='inner'
)

print(f"\nüìä R√©sultats des strat√©gies de mapping:")
print(f"   Strat√©gie A (nom fichier ‚Üí requ√™te): {len(data_a)} matchs")
print(f"   Strat√©gie B (num√©ro carte ‚Üí requ√™te): {len(data_b)} matchs")
print(f"   Strat√©gie C (num√©ro permis ‚Üí requ√™te): {len(data_c)} matchs")

# Choisir la meilleure strat√©gie
strategies = [
    ('A - Nom fichier', data_a),
    ('B - Num√©ro carte', data_b),
    ('C - Num√©ro permis', data_c)
]

best_strategy = max(strategies, key=lambda x: len(x[1]))
strategy_name, data = best_strategy

print(f"\n‚úÖ Meilleure strat√©gie: {strategy_name} ({len(data)} matchs)")

# Filtrer les donn√©es valides
# Garder seulement les lignes avec un nom d'examen valide (>5 caract√®res)
data_clean = data[data['exam_name'].str.len() > 5].copy()

print(f"\nüìà Dataset apr√®s filtrage:")
print(f"   Total apr√®s fusion: {len(data)} exemples")
print(f"   Avec exam_name valide (>5 chars): {len(data_clean)} exemples")

if len(data_clean) < 20:
    print("\n‚ö†Ô∏è  ATTENTION: Moins de 20 exemples valides!")
    print("   L'OCR n'a peut-√™tre pas bien fonctionn√©.")
    print("   V√©rifiez le fichier ocr_results.csv pour voir les textes extraits.")
else:
    print(f"\n‚úÖ Dataset suffisant pour l'entra√Ænement!")

# Afficher un √©chantillon
print("\nüìã √âchantillon du dataset final:")
print(data_clean[['image_name', 'exam_name', 'protocol']].head(15))

# Distribution des protocoles dans le dataset
print("\nüìä Distribution dans le dataset final:")
print(data_clean['protocol'].value_counts())

# Sauvegarder le dataset complet
dataset_path = os.path.join(OUTPUT_DIR, 'dataset_complet.csv')
data_clean.to_csv(dataset_path, index=False)
print(f"\nüíæ Dataset sauvegard√©: {dataset_path}")
````

````python
import pandas as pd
import re

# Nettoyage simple des noms d'examen
def clean_text(text):
    text = text.lower()  # minuscule
    text = re.sub(r'[^a-z0-9√†√¢√©√®√™√¥√π√ß\s]', ' ', text)  # garder lettres, chiffres et accents
    text = re.sub(r'\s+', ' ', text).strip()  # supprimer espaces multiples
    return text

data_clean['exam_name_clean'] = data_clean['exam_name'].apply(clean_text)
````
